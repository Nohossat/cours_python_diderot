#### Q1 - Créer une fonction `calc` avec la signature suivante : `calc(terme1, operation, terme2)`, `terme1` et `terme2` sont des nombres (int, float acceptés) et `operation` un des symboles suivants : "+", "-", "/", "*". La variable `operation` permettra au programme d'identifier le calcul à réaliser. Par exemple, `calc(2, "+", 3)` retournera 5.

```python
def calc(terme1, operation, terme2):
    
    def add(x, y):
        return x + y
    
    def substract(x, y):
        return x - y
    
    def multiply(x, y):
        return x * y
    
    def divide(x, y):
        return x / y
        
    operations = {
        "+" : add,
        "-": substract,
        "*": multiply,
        "/": divide
    }
    
    result = operations[operation](terme1, terme2)
    return result
```

#### Q2 - On va améliorer cette fonction avec la validation des arguments. Pour rappel, `terme1` et `terme2` n'acceptent que des nombres, `operation` est une chaîne de caractères ne pouvant être qu'un des symboles décrits dans Q1.

```python
def add(x, y):
    return x + y
    
def substract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

def calc(terme1, operation, terme2):    
    operations = {
        "+" : add,
        "-": substract,
        "*": multiply,
        "/": divide
    }
    
    if not isinstance(terme1, (int, float)) or not isinstance(terme2, (int, float)):
        return 'Seules des valeurs numériques sont acceptées pour les variables terme1 et terme2'
    
    if operation not in operations.keys():
        return 'Les opérations possibles sont : +, -, *, /'
    
    result = operations[operation](terme1, terme2)
    return result
```

#### Q3 - Maintenant que notre calculatrice est opérationnelle, on souhaite mettre en place une interface utilisateur. Créer une autre fonction `interface_calc` qui permettra de demander à un utilisateur, les termes et le type d'opération qu'ils souhaite réaliser. N'oubliez pas le contrôle sur les données récoltées avant de passer ces valeurs à la fonction `calc`. 

```python
def validate_input(idx_terme):
        nb = ""
        flag_start_nb = True 
        while not nb.isnumeric():
            if flag_start_nb:
                flag_start_nb = False
            else :
                print(f"Le terme {idx_terme} doit être une valeur numérique")

            nb = input(f"Terme{idx_terme} :")
        
        return nb
    
    
def validate_operation():
    op = ""
    flag_start_op = True 

    while op not in ["+", "-", "*", "/"]:
        if flag_start_op:
            flag_start_op = False
        else :
            print("Erreur : les opérations possibles sont +, -, *, /")

        op = input("Opération souhaitée, les opérations possibles sont +, -, *, /:")

    return op


def interface_calc():
    # ici on va utiliser la fonction validate_input pour valider les valeurs passées par l'utilisateur
    # on utilise une boucle pour appliquer cette validation sur les deux termes à récupérer
    # l'unpacking permet d'affecter les valeurs de la compréhension de liste à deux variables nb1 et nb2. Ce sont les variables que l'on passera 
    # à la calculatrice in fine
    nb1, nb2 = [float(validate_input(i)) for i in [1, 2]]

    op = validate_operation()
    
    return calc(nb1, op, nb2)
```

#### Q4 - Bonus - Transformer ce programme en une classe nommée Calculatrice

```python
class Calculatrice:
    def __init__(self, terme1 = "", terme2 = "", op = ""):
        self.terme1 = terme1
        self.terme2 = terme2
        self.op = op
        
        
    def add(self):
        return self.terme1 + self.terme2

    
    def substract(self):
        return self.terme1 - self.terme2

    
    def multiply(self):
        return self.terme1 * self.terme2

    
    def divide(self):
        return self.terme1 / self.terme2
        
        
    @staticmethod
    def validate_input():
        nb = ""
        flag_start_nb = True 
        while not nb.isnumeric():
            if flag_start_nb:
                flag_start_nb = False
            else :
                print(f"Le terme {idx_terme} doit être une valeur numérique")

            nb = input(f"Terme{idx_terme} :")
        
        return nb
    
    
    def validate_operation(self):
        flag_start_op = True 

        while self.op not in ["+", "-", "*", "/"]:
            if flag_start_op:
                flag_start_op = False
            else :
                print("Erreur : les opérations possibles sont +, -, *, /")

            self.op = input("Opération souhaitée, les opérations possibles sont +, -, *, /:")
    
    
    def __calc(self): # we make this function private because we don't want it to be used by itself
        operations = {
            "+": self.add,
            "-": self.substract,
            "*": self.multiply,
            "/": self.divide
        }
        
        # plus besoin de faire la validation dans __calc puisque on fait la validation des valeurs dans interface_calc
        result = operations[self.op]()
        return result
    
    
    def interface_calc(self):
        self.terme1, self.terme2 = [float(validate_input(i)) for i in [1, 2]]
        self.validate_operation()

        return self.__calc()
```

### Exercice 3 - le jeu du Morpion

Nous allons essayer de reproduire un morpion en Python. Pour des raisons de simplicité, chaque case sera identifiée par un numéro entre 0 et 8. Comme ci-dessous:

0 | 1 | 2  
3 | 4 | 5

#### Q1 - La ligne `(0, 1, 2)` représente une combinaison gagnante. En effet, le jeu s'arrête si cette ligne est remplie de 'O' ou de 'X'. Identifier toutes les combinaisons gagnantes d'un morpion. Pour info, il y en a 8. Chaque combinaison sera un tuple comme par exemple `(0, 1, 2)`. On les ajoutera à une liste nommée `combinaisons_gagnantes`.

```python
combinaisons_gagnantes = [
    (0, 1, 2), 
    (3, 4, 5),
    (6, 7, 8),
    (0, 3, 6),
    (1, 4, 7),
    (2, 5, 8),
    (0, 4, 8),
    (2, 4, 6)
]
```

#### Q2 - Créer une fonction qui permet d'afficher un plateau de morpion. Cette fonction s'appelera `print_board`, elle prendra en argument une liste `board` de 9 chaînes de caractères. Pour le moment, ces chaînes de caractères seront toutes égales à ".". Cette fonction permet donc de modifier l'affichage d'une liste : au lieu d'une ligne de 1x9, nous souhaitons un affichage de 3x3, comme ci-dessous:

| . | . | . |  
| . | . | . |  
| . | . | . |  


```python
board = ["."] * 9

def print_board(board):
    morpion = ""
    for idx, case in enumerate(board):
        morpion = morpion + "|"
        morpion = morpion + str(case)
        
        if (idx + 1) % 3 == 0:
            morpion = morpion + "|\n"
    
    print(morpion)
    
print_board(board)
```

#### Q3 - Créer une fonction `update_board` prenant en argument un nombre `idx` entre 0 et 8, une chaîne de caractères `symbol` et la liste `board`. Cette fonction nous permettra de mettre à jour `board` : on remplacera la valeur à l'indice `idx` par la valeur contenue dans `symbol`. Par exemple, si nous souhaitons mette un 'O' à l'indice 2 du `board`, on appelera `update_board(2, 'O', board)`.

```python
def update_board(idx, symbol, board):
    new_board = board.copy()
    new_board[idx] = symbol
    return new_board
```

```python
update_board(2, "O", board)
```

#### Q4 - Créer une fonction `check_if_case_available` : cette fonction permet de vérifier si la case à l'indice `idx` est vide ou non. Retourner `True` si la case est vide. Cette petite fonction nous sera utile pour éviter d'écraser la valeur d'une case non vide. Elle prend `idx` et `board` comme arguments.

```python
def check_if_case_available(idx, board):
    return board[idx] == "."
```

#### Q5 - On commence à avoir toutes les pièces nécessaires pour mettre en place ce jeu. Le morpion se joue à 2. Créer une fonction `start_game`. Cette fonction réalise plusieurs actions.

Dans un premier temps, on initialise le jeu : 
- Elle prend en argument un liste `joueurs` ayant pour éléments "O" et "X". Un joueur est identifié par son symbole.
- Elle crée un plateau `board` => une liste de 9 éléments tous ayant la valeur "." par défaut
- Elle crée la variable `joueur_actuel` qui par défaut sera le premier élément de la liste `joueurs`

Puis on utilise une boucle `while` permettant d'alterner entre joueur O et joueur X. La boucle commencera par `while True:`. Nous verrons la condition d'arrêt à la question suivante.

Dans cette boucle on décrit le déroulement d'un tour : 
- on affiche le plateau
- on demande au joueur actuel de choisir la case à cocher (une case étant identifiée par son indice)
- vérifier que la case sélectionnée est vide : tant que cette condition n'est pas vérifiée, on demandera au joueur de choisir à nouveau
- on actualise le plateau en conséquence
- on change de joueur

```python
def start_game(joueurs=['O', 'X']):
    board = ["."] * 9
    idx_joueur = 0
    joueur_actuel = joueurs[idx_joueur]
   
    while True:
        # on affiche le plateau
        print_board(board)
        
        # on demande au joueur l'indice d'une case
        case_idx = input(f"Joueur {joueur_actuel}, à ton tour: Indiquer la case à cocher : ")
        
        # on vérifie que la case est vide
        while not check_if_case_available(int(case_idx), board):
            case_idx = input(f"Joueur {joueur_actuel}, la case n'est pas vide, veuillez en choisir une autre: ")
            
        # on met à jour le plateau
        board = update_board(int(case_idx), joueur_actuel, board)
        
        # on passe la main au joueur suivant
        idx_joueur = (idx_joueur + 1) % 2
        joueur_actuel = joueurs[idx_joueur]
    
start_game(joueurs=['O', 'X'])
```

#### Q6 - Nous avons vu qu'avec `while True`, nous avions une boucle infinie. Mais nous savons que le jeu s'arrête lorsqu'un des deux joueurs réalise une combinaison gagnante. Il faudra donc inspecter le plateau après chaque coup pour voir si un des deux joueurs a gagné.

Créer une fonction `check_winner` qui prendra en argument `board`, `symbol` et les combinaisons gagnantes définies à la question 1.

La liste `combinaisons_gagnantes` identifie les indices des cases permettant au joueur O ou X de gagner. Nous pouvons déclarer la victoire si toutes les cases d'une combinaison gagnante ont le même symbole. Retourner `True` si une combinaison gagnante a été identifiée dans le plateau sinon `False`.

On peut aussi ajouter une condition sur le remplissage du plateau : si après vérification des combinaisons gagnantes, on constate qu'il n'y a plus de case disponible, on annonce la fin du jeu : il n'y a pas de gagnant.

N'oubliez pas de tester votre fonction.

```python
def check_winner(board, joueur_actuel, combinaisons_gagnantes):
    for combi in combinaisons_gagnantes:
        idx1, idx2, idx3 = combi
        if board[idx1] == board[idx2] == board[idx3] == joueur_actuel:
            return True
    
    return False
```

```python
board = ["O", "O", ".", ".", "O", ".", ".", "O", "."]
joueur_actuel = "O"
check_winner(board, joueur_actuel, combinaisons_gagnantes)
```

```python
def is_board_full(board):
    if "." not in board:
        return True
    
    return False
```

```python
def start_game(joueurs=['O', 'X']):
    board = ["."] * 9
    idx_joueur = 0
    joueur_actuel = joueurs[idx_joueur]
    res = ""
    
    combi_gagnantes = [
        (0, 1, 2), 
        (3, 4, 5),
        (6, 7, 8),
        (0, 3, 6),
        (1, 4, 7),
        (2, 5, 8),
        (0, 4, 8),
        (2, 4, 6)
    ]
   
    while True:
        print_board(board)
        case_idx = input(f"Joueur {joueur_actuel}, à ton tour: Indiquer la case à cocher : ")
        
        # on vérifie que la case est vide
        while not check_if_case_available(int(case_idx), board):
            case_idx = input(f"Joueur {joueur_actuel}, la case n'est pas vide, veuillez en choisir une autre: ")
            
        # on met à jour le plateau
        board = update_board(int(case_idx), joueur_actuel, board)
        
        # on vérifie si ce coup est gagnant
        if check_winner(board, joueur_actuel, combi_gagnantes):
            res = f"Joueur {joueur_actuel} gagne la partie"
            break
        
        # on vérifie s'il s'agit du dernier coup de la partie
        if is_board_full(board):
            res = "Le jeu est terminé. Il n'y a plus de cases disponibles."
            break
        
        # on passe la main au joueur suivant
        idx_joueur = (idx_joueur + 1) % 2
        joueur_actuel = joueurs[idx_joueur]
    
    # le jeu est terminé
    print_board(board)
    print(res)
    return None
        
        
start_game(joueurs=['O', 'X'])
```
